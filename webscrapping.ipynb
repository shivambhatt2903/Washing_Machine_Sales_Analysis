## Import Libraries

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from itertools import zip_longest
```

## URL Of Flipkart Website

```python
url="https://www.flipkart.com/search?q=washing+machines&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
```

## Headers User-Agent

```python
Headers=({"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36"})
```

## Sending Request To The Website

```python
webpage=requests.get(url,headers=HEADERS)
```

```python
webpage
```

## Using Beautiful Soup 

```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

## Scrap The 31 Pages Information of Flipkart Product

```python
informations=[]
for i in range(1,32):
    url=f"https://www.flipkart.com/search?q=washing+machines&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    webpage=requests.get(url,headers=HEADERS)
    soup=BeautifulSoup(webpage.content,"html.parser")
    info=soup.find_all("div",class_="KzDlHZ")
    for infos in info:
        informations.append(infos.text)
print(informations)
```

## Scrap The Speed Rating And Warranty OF Product

```python
Max_Speed=[]
Electric_Consumption_Rating=[]
Warranty=[]
for i in range(1,32):
    url=f"https://www.flipkart.com/search?q=washing+machines&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    webpage=requests.get(url,headers=HEADERS)
    soup=BeautifulSoup(webpage.content,"html.parser")
    all_info=soup.find_all("li",class_="J+igdf")
    for item in all_info:
        text = item.get_text(strip=True)
        if "rpm Max Speed" in text:
            Max_Speed.append(text) 
        elif "Star Rating" in text:
            Electric_Consumption_Rating.append(text)
        elif "Warranty" in text:
            Warranty.append(text)
print("Max_Speed :" , Max_Speed)
print("Electric_Consumption_Rating :",Electric_Consumption_Rating)
print("Warranty :" , Warranty)
        
        
    
    
    
```

## Scrap The Prices Of Product

```python
Prices=[]
for i in range(1,32):
    url=f"https://www.flipkart.com/search?q=washing+machines&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    webpage=requests.get(url,headers=HEADERS)
    soup=BeautifulSoup(webpage.content,"html.parser")
    amount=soup.find_all("div","Nx9bqj _4b5DiR")
    for p in amount:
        Prices.append(p.text)
print("Prices :",Prices)
    
```

## Scrap The Discount% Of Product

```python
Discount_Percentage=[]
for i in range(1,32):
    url=f"https://www.flipkart.com/search?q=washing+machines&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    webpage=requests.get(url,headers=HEADERS)
    soup=BeautifulSoup(webpage.content,"html.parser")
    dis=soup.find_all("div",class_="UkUFwK")
    for d in dis:
        Discount_Percentage.append(d.text)
print("Discount Percentage :",Discount_Percentage)
```

## Combined All Data To Make DataFrame

```python
combined = list(zip_longest(informations, Max_Speed, Electric_Consumption_Rating, Warranty, Prices, Discount_Percentage, fillvalue=None))
df = pd.DataFrame(combined, columns=["Informations", "Max_Speed", "Electric_Consumption_Rating", "Warranty", "Prices", "Discount_Percentage"])
df.head()
```

## Export The Scrap Data In CSV Format

```python
df.to_csv("flipkart_washing_machine.csv")
```
